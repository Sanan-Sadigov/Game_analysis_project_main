```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df1=pd.read_csv('df1.csv',parse_dates=['eventTimestamp'])
df1.drop_duplicates(inplace=True)
df1['eventTimestamp']=df1['eventTimestamp'].str.slice(0,10)
df1['eventTimestamp']=pd.to_datetime(df1['eventTimestamp'],format='mixed')
```

• For every day and every platform: The total of unique users connected on that day. Name this column “DAU”.

```python
df1=df1[df1['eventName']=='gameStarted']
Data=df1.groupby(['platform','eventName','eventTimestamp'])['userID'].nunique().reset_index()
Data=Data.rename(columns={'userID': 'DAU'})
Data
```

• The total of new unique users who entered the app on that day. Name this column “New Players”. 

```python
df1=df1[df1['eventName']=='gameStarted']
df1=df1[df1['eventTimestamp']==df1['gaUserStartDate']]
df1[['eventTimestamp','eventName','userID']]
new_unique=df1.groupby(['platform','eventName','eventTimestamp'])['userID'].nunique().reset_index()
new_unique=new_unique.rename(columns={'userID': 'New Players'})
new_unique.sort_values(by='New Players', ascending=False)
```

• The raw total of sessions (a session is when the game is started). Name this column “Sessions”.  

```python
sess=df1[df1['eventName']=='gameStarted']
sess=df1.groupby(['eventTimestamp','platform'])['userID'].count().reset_index()
sess=sess.rename(columns={'eventName': 'Sessions'})
sess
```

• The raw total of matches started. Name this column “Matches”.  

```python
df1=pd.read_csv('df1.csv',parse_dates=['eventTimestamp'])
df1.drop_duplicates(inplace=True)
df1['eventTimestamp']=df1['eventTimestamp'].str.slice(0,10)
df1['eventTimestamp']=pd.to_datetime(df1['eventTimestamp'],format='mixed')
df1=df1[df1['eventName']=='matchStarted']
Matches=df1.groupby(['eventTimestamp','platform'])['userID'].size().reset_index(name='Matches')
Matches
```

• The raw total of matches lost. Name this column “Losses”.  

```python
df1=pd.read_csv('df1.csv',parse_dates=['eventTimestamp'])
df1.drop_duplicates(inplace=True)
df1['eventTimestamp']=df1['eventTimestamp'].str.slice(0,10)
df1['eventTimestamp']=pd.to_datetime(df1['eventTimestamp'],format='mixed')
df1=df1[df1['matchResult']=='Loss']
Losses=df1.groupby(['eventTimestamp','platform'])['userID'].size().reset_index(name='Losses')
Losses
```

• The raw total of matches won. Name this column “Wins”.  

```python
df1=pd.read_csv('df1.csv',parse_dates=['eventTimestamp'])
df1.drop_duplicates(inplace=True)
df1['eventTimestamp']=df1['eventTimestamp'].str.slice(0,10)
df1['eventTimestamp']=pd.to_datetime(df1['eventTimestamp'],format='mixed')
df1=df1[df1['matchResult']=='Win']
Win=df1.groupby(['eventTimestamp','platform'])['userID'].size().reset_index(name='Wins')
Win
```
